# 环境配置
**Setup**

在本节中，我们将介绍 XiangShan 所需的环境配置方法。如果您正在 tutorial 的现场阅读此 Notebook，您可以跳过此节，因为我们已经帮您配置好了环境。

In this section, we will introduce the setup steps of preparing environment required by XiangShan. If you are reading this Notebook on-site of the tutorial, you can skip this section as we have already set up the environment for you.

## Bootcamp

一个快速体验 XiangShan 的开发流程和工具链的好方法是使用我们提供的 [Bootcamp](https://github.com/OpenXiangShan/bootcamp) 仓库。您正在阅读的这个 Notebook 就是 Bootcamp 的一部分。

The Bootcamp repository is a good way to quickly experience the development workflow and toolchain of XiangShan. The Notebook you are reading is part of the Bootcamp.

在 Bootcamp 中配置环境非常简单，您只需：
1. 运行 `git clone https://github.com/OpenXiangShan/bootcamp.git` 克隆 Bootcamp 仓库到本地
2. 运行 `git submodule update --init --recursive` 初始化子模块
3. 运行 `docker compose up -d` 启动 Bootcamp 容器
4. 一切就绪，前往 `http://localhost:8443` 打开 code-server，开始您的 XiangShan 之旅！

The environment setup in Bootcamp is straightforward. You only need to:
1. Run `git clone https://github.com/OpenXiangShan/bootcamp.git` to clone the Bootcamp repository locally
2. Run `git submodule update --init --recursive` to initialize submodules
3. Run `docker compose up -d` to start the Bootcamp container
4. Everything is ready! Go to `http://localhost:8443` to open code-server and start your XiangShan journey!

## xs-env

Bootcamp 仅适用于想要快速体验 XiangShan 的用户，如果您需要进行更深入的开发，建议您使用 [xs-env](https://github.com/OpenXiangShan/xs-env) 来配置您的开发环境。

Bootcamp is only suitable for users who want to quickly experience XiangShan. If you need to do more in-depth development, it is recommended to use [xs-env](https://github.com/OpenXiangShan/xs-env) to configure your development environment.

请查阅[文档](https://docs.xiangshan.cc/zh-cn/latest/tools/xsenv)了解如何使用 xs-env 来配置您的开发环境。

Please refer to [our documentation site](https://docs.xiangshan.cc/zh-cn/latest/tools/xsenv-en/) to learn how to use xs-env to set up your development environment.
